## Calculate ROC Curves for groups of interest (Arabic)
### Author: Ben Greenawald

```python
import os

from sklearn import metrics
```

```python
# Read in the document labels
group_labels = {}
with open("arabic-groups-labels.txt", 'r') as file:
    for line in file.readlines():
        split = line.split(",")
        group_name = split[0].replace(" ", "-")
        group_labels[group_name] = int(split[1])
        file.close()
```

# Logistic Regression

```python
true_label = []
probs = []
files = []

for filename in os.listdir("prob_LR/"):
    with open("prob_LR/" + filename, 'r') as file:
        for line in file.readlines():
            split = line.split(",")
            probs.append(float(split[1]))
            group_name = split[0].split("_g")[0]
            true_label.append(int(group_labels[group_name]))
            files.append(split[0].strip())
```

```python
fpr_lr, tpr_lr, threshold = metrics.roc_curve(true_label, probs)
roc_auc_lr = metrics.auc(fpr_lr, tpr_lr)
```

# CNN

```python
true_label_cnn = []
probs_cnn = []
seen = []

with open("ar_cnn_log_logocv_aug_prob.txt", 'r') as file:
    for line in file.readlines():
        if line.startswith("group"):
            continue
        split = line.split(" ")
        name = split[0].split('.')[0]
        if name in files:
            probs_cnn.append(float(split[1]))
            group_name = name.split("_g")[0]
            true_label_cnn.append(int(group_labels[group_name]))
            seen.append(name)
```

```python
fpr_cnn, tpr_cnn, threshold = metrics.roc_curve(true_label_cnn, probs_cnn)
roc_auc_cnn = metrics.auc(fpr_cnn, tpr_cnn)
```

```python
seen[-10:]
```

# LSTM

```python
true_label_lstm = []
probs_lstm = []
seen = []
with open("ar_lstm_log_logocv_aug_prob.txt", 'r') as file:
    for line in file.readlines():
        if line.startswith("group"):
            continue
        split = line.split(" ")
        name = split[0].split('.')[0].strip()
        if name in files:
            probs_lstm.append(float(split[1]))
            group_name = name.split("_g")[0]
            true_label_lstm.append(int(group_labels[group_name]))
            seen.append(name)
```

```python
fpr_lstm, tpr_lstm, threshold = metrics.roc_curve(true_label_lstm, probs_lstm)
roc_auc_lstm = metrics.auc(fpr_lstm, tpr_lstm)
```

# Plot

```python
# method I: plt
import matplotlib.pyplot as plt
plt.title('ROC for High Performance Models (Arabic)')
plt.plot(fpr_lr, tpr_lr, 'b', label = 'Logistic Regression: AUC = %0.2f' % roc_auc_lr)
plt.plot(fpr_cnn, tpr_cnn, 'y', label = 'CNN: AUC = %0.2f' % roc_auc_cnn)
plt.plot(fpr_lstm, tpr_lstm, 'g', label = 'LSTM: AUC = %0.2f' % roc_auc_lstm)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.savefig("ROC (arabic).png")
plt.show()
```
